```python
import pandas as pd
df_sales = pd.read_excel('/content/Metro sales summary.xlsx')
df_sales
```

```python
df_sales['SPAT without GST'] = df_sales['SPAT']/(1+(df_sales['GST %']/100))
df_sales.head()
```

```python
df_sales['Sales'] = df_sales['SPAT without GST']*df_sales['Metro Sale Qty']
df_sales['Gross Profit'] = (df_sales['SPAT without GST']-df_sales['Rate'])*df_sales['Metro Sale Qty']
df_sales.head()
```

```python
df_sales['Month'].unique()
```

```python
df_sales['Month'] = df_sales['Month'].replace('MAY','May')
df_sales.head()

```

```python
df_sales['Month'].unique()
```

```python
month_dict = {
    'Jan': '01', 'Feb': '02', 'Mar':'03', 'April': '04',
    'May': '05', 'June': '06', 'July': '07', 'Aug': '08',
    'Sep': '09', 'Oct': '10', 'Nov': '11', 'Dec': '12'
}

df_sales['Date'] = df_sales.apply(lambda row: f"{str(row['Year'])[-2:]}/{month_dict[row['Month']]}", axis=1)


df_sales['Date'] = pd.to_datetime(df_sales['Date'], format='%y/%m')


df_sales = df_sales.sort_values(by='Date')


df_sales['Date'] = df_sales['Date'].dt.strftime('%y/%m')


df_sales = df_sales.reset_index(drop=True)

df_sales['Date'].unique()
```

```python
df_sales['Profit per Cubic Metre'] = df_sales['Gross Profit']/df_sales['CBM']
df_sales.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

profit_by_month_category = df_sales.groupby(['Date', 'Category'])['Profit per Cubic Metre'].mean().reset_index()


plt.figure(figsize=(12, 6))
sns.lineplot(data=profit_by_month_category, x='Date', y='Profit per Cubic Metre', hue='Category')

plt.xlabel('Date')
plt.xticks(rotation=45)


plt.ylabel('Profit per Cubic Metre')
plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))


plt.title('Profit per Cubic Metre Over Time by Category')


plt.tight_layout()


plt.show()

```

```python

import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

profit_sales_by_date = df_sales.groupby('Date')[['Gross Profit', 'Sales']].sum().reset_index()

plt.figure(figsize=(12, 6))
bars1 = plt.bar(profit_sales_by_date['Date'], profit_sales_by_date['Sales'], label='Sales')
bars2 = plt.bar(profit_sales_by_date['Date'], profit_sales_by_date['Gross Profit'], bottom=profit_sales_by_date['Sales'], label='Gross Profit')


for bar in bars1:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2.0, height, f'{height:,.0f}', ha='center', va='bottom')

for bar in bars2:
    height = bar.get_height()
    bottom = bar.get_y() + height
    plt.text(bar.get_x() + bar.get_width() / 2.0, bottom, f'{height:,.0f}', ha='center', va='bottom')

plt.xlabel('Date')
plt.xticks(rotation=45)

plt.ylabel('Value')
plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))

plt.title('Gross Profit and Sales by Date')
plt.legend()

plt.tight_layout()

plt.show()

```

```python

import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

category_profit_sales = df_sales.groupby('Category')[['Gross Profit', 'Sales']].sum().reset_index()

plt.figure(figsize=(12, 6))
plt.bar(category_profit_sales['Category'], category_profit_sales['Sales'], label='Sales')
plt.bar(category_profit_sales['Category'], category_profit_sales['Gross Profit'], bottom=category_profit_sales['Sales'], label='Gross Profit')

plt.xlabel('Category')
plt.ylabel('Value')

plt.title('Gross Profit and Sales by Category')

plt.legend()

plt.xticks(rotation=45, ha='right')
plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))

plt.tight_layout()

plt.show()

```

```python

import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

df_2023 = df_sales[df_sales['Date'].str.startswith('23')]
df_2024 = df_sales[df_sales['Date'].str.startswith('24')]

monthly_2023 = df_2023.groupby(df_2023['Date'].str[-2:])[['Sales', 'Gross Profit']].sum().reset_index()
monthly_2024 = df_2024.groupby(df_2024['Date'].str[-2:])[['Sales', 'Gross Profit']].sum().reset_index()

plt.figure(figsize=(12, 6))

plt.plot(monthly_2023['Date'], monthly_2023['Sales'], label='Sales 2023', marker='o')
plt.plot(monthly_2023['Date'], monthly_2023['Gross Profit'], label='Gross Profit 2023', marker='o')

plt.plot(monthly_2024['Date'], monthly_2024['Sales'], label='Sales 2024', marker='o')
plt.plot(monthly_2024['Date'], monthly_2024['Gross Profit'], label='Gross Profit 2024', marker='o')

plt.xlabel('Month')
plt.ylabel('Value')
plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))
plt.title('Monthly Sales and Gross Profit for 2023 and 2024')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
